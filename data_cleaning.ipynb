```python
import pandas as pd

df = pd.read_csv("data/udemy_lang.csv")
```

```python
df.shape
```

```python
df = df[["title", "url", "duration", "language"]]
```

```python
df = df.dropna(subset=["title", "url"])
```

```python
df["title"] = df["title"].str.lower().str.strip()
df["language"] = df["language"].fillna("unknown").str.lower()
```

```python
df.isnull().sum()
df.shape
```

```python
domain_keywords = {
    "AI": [
        "artificial intelligence",
        "machine learning",
        "deep learning",
        "neural network",
        "computer vision",
        "nlp",
        " ai ",
        "ai-",
        " ai"
    ],
    "Web": ["web development", "html", "css", "javascript", "react", "angular"],
    "Data": ["data science", "data analysis", "pandas", "numpy", "sql"],
    "Cloud": ["aws", "azure", "google cloud", "cloud computing", "devops"]
}
```

```python
def assign_domain(title):
    title = " " + title + " "   # avoids ai inside agile
    for domain, keywords in domain_keywords.items():
        for kw in keywords:
            if kw in title:
                return domain
    return "Other"
```

```python
df["domain"] = df["title"].apply(assign_domain)
```

```python
df["domain"].value_counts()
```
