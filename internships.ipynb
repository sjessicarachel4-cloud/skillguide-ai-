```python
import pandas as pd
df = pd.read_csv("data/internships.csv")
```

```python
df.columns
```

```python
df = df[[
    "Job Title",
    "Location",
    "Experience",
    "Skill 1", "Skill 2", "Skill 3", "Skill 4", "Skill 5", "Skill 6"
]]
```

```python
df = df.dropna(subset=["Job Title"])
```

```python
df["Job Title"] = df["Job Title"].str.lower().str.strip()
df["Location"] = df["Location"].str.lower().str.strip()
```

```python
skill_cols = ["Skill 1", "Skill 2", "Skill 3", "Skill 4", "Skill 5", "Skill 6"]

df["skills"] = df[skill_cols].fillna("").agg(", ".join, axis=1)

df["skills"] = df["skills"].str.lower()
```

```python
df = df.drop_duplicates()
```

```python
skill_cols = ["Skill 1", "Skill 2", "Skill 3", "Skill 4", "Skill 5", "Skill 6"]

df["Skills"] = (
    df[skill_cols]
    .fillna("")                 # remove NaN
    .agg(", ".join, axis=1)     # join skills
    .str.replace(r"(, )+", ", ", regex=True)
    .str.strip(", ")
)
```

```python
df.drop(columns=skill_cols, inplace=True)
```

```python
def assign_domain(text):
    text = str(text).lower()

    ai_keywords = [
        "machine learning", "deep learning", "artificial intelligence",
        "nlp", "computer vision", "ml", "ai"
    ]

    data_keywords = [
        "data analyst", "data science", "data scientist",
        "data analysis", "analytics", "big data"
    ]

    web_keywords = [
        "web", "frontend", "backend", "full stack",
        "react", "angular", "html", "css", "javascript"
    ]

    cloud_keywords = [
        "cloud", "aws", "azure", "gcp", "devops", "kubernetes", "docker"
    ]

    for kw in ai_keywords:
        if kw in text:
            return "AI"

    for kw in data_keywords:
        if kw in text:
            return "Data"

    for kw in web_keywords:
        if kw in text:
            return "Web"

    for kw in cloud_keywords:
        if kw in text:
            return "Cloud"

    return "Other"


# apply on Job Title + Skills together (best practice)
df["domain"] = (
    df["Job Title"].astype(str) + " " + df["Skills"].astype(str)
).apply(assign_domain)

df.head()
```
